```python
import pytz
from datetime import datetime

def convert_to_utc(timezone,start_time_local,end_time_local):

# Define the "America/Chicago" timezone
  timezone = pytz.timezone(timezone)

# Define the local times
  start_time_local = datetime.strptime(start_time_local, '%H:%M:%S')
  end_time_local = datetime.strptime(end_time_local, '%H:%M:%S')

  # Localize the local times to the "America/Chicago" timezone
  start_time_chicago = timezone.localize(start_time_local)
  end_time_chicago = timezone.localize(end_time_local)

  # Convert to UTC
  start_time_utc = start_time_chicago.astimezone(pytz.utc)
  end_time_utc = end_time_chicago.astimezone(pytz.utc)

  # Format the result as strings
  start_time_utc_str = start_time_utc.strftime('%H:%M:%S')
  end_time_utc_str = end_time_utc.strftime('%H:%M:%S')



  return [start_time_utc_str,end_time_utc_str]
```
